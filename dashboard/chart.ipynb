```python
from pymongo import MongoClient
import matplotlib.pyplot as plt
```

```python
client = MongoClient('mongodb://localhost:27017/')
db = client['datakelahiranDB']
collection1 = db['dashboard_data2018']
```

```python
 data = collection1.find_one()
 print(data)
```

```python
x = []
y1 = []
y2 = []

for document in data:
    x.append(document['wilayah'])
    y1.append(document['jumlah_bayi_lahir'])
    y2.append(document['jumlah'])
```

```python
# Menentukan lebar batang
bar_width = 0.2

# Menentukan posisi bar pada sumbu x
bar1 = [i - 1.5 * bar_width for i in range(len(x))]
bar2 = [i - 0.5 * bar_width for i in range(len(x))]

# Membuat grafik batang
plt.bar(bar1, y1, width=bar_width, label='Jumlah Bayi Lahir')
plt.bar(bar2, y2, width=bar_width, label='Jumlah Sesuai Kondisi')

# Memberikan label sumbu x dan y
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

# Memberikan label pada tiap bar
plt.xticks([i for i in range(len(x))], x)

# Menampilkan legenda
plt.legend()

# Menampilkan grafik batang
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data untuk sumbu x dan y
x = ['A', 'B', 'C', 'D']
y1 = [10, 20, 15, 25]
y2 = [5, 15, 10, 20]
kondisi = ['Kondisi 1', 'Kondisi 2', 'Kondisi 1', 'Kondisi 2']

# Menentukan lebar batang
bar_width = 0.4

# Menentukan posisi bar pada sumbu x
bar1 = [i - 0.2 for i in range(len(x))]
bar2 = [i + 0.2 for i in range(len(x))]

# Membuat grafik batang
plt.bar(bar1, y1, width=bar_width, label='Field 1')
plt.bar(bar2, y2, width=bar_width, label='Field 2')

# Memberikan label sumbu x dan y
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

# Memberikan label pada tiap bar
plt.xticks(range(len(x)), x)

# Menambahkan kondisi pada setiap bar pada field 2
for i in range(len(x)):
    plt.text(bar2[i], y2[i], kondisi[i], ha='center')

# Menampilkan legenda
plt.legend()

# Menampilkan grafik batang
plt.show()
```
